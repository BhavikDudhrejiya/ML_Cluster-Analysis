# Basic Analysis of the data

```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#View full view of the data
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Loading the data
data = pd.read_csv('C:\\Users\\bhavi\\OneDrive\\Desktop\\Data\\Wholesale customers data.csv')
```

```python
#View the top five rows of the data
data.head()
```

# Normalization of the data

```python
#Normalization of the data
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
#This normalized data are created in numpy array which needs to be converted into data frame
```

# Normalized data converted into data frame

```python
#Numpy array data converted into dataframe
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)

#Checking top five rows
data_scaled.head()
```

# Hierarchical clustering - Agglomerative method

```python
#Importing library for hierarchical clustering
import scipy.cluster.hierarchy as shc
```

```python
#Creating Dendrograms 
plt.figure(figsize=(16,9)) #Setting size of the plot
plt.title('Dendrograms') #Titel of the plot
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward')) #single, complete, average, weighted, centroid, median, ward 
```

```python
#Creating Dendrograms
plt.figure(figsize=(16,9))
plt.title('Dendrograms')
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y = 6, color = 'r', linestyle = '--')
plt.show()
```

```python
#Importing library
from sklearn.cluster import AgglomerativeClustering

#Creating Agglomerative Clusters
cluster = AgglomerativeClustering(n_clusters=2, 
                                  affinity='euclidean', 
                                  linkage='ward')

cluster.fit_predict(data_scaled)
```

```python
#Plotting the Milk and Grocery
plt.figure(figsize=(10,7))
plt.scatter(data_scaled['Milk'], 
            data_scaled['Grocery'], 
            c = cluster.labels_) #Coloring the data as per cluster lables
plt.show()
```
